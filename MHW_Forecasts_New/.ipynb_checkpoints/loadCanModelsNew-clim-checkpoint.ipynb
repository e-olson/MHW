```python
# ensemble members 1-20 = GEM5.2-NEMO
# ensemble members 21-40 = CanESM5
GN=f1.sst.isel(number=slice(0,20))
C5=f1.sst.isel(number=slice(20,40))
```

```python
GN_Emean=GN.mean(dim='number',keepdims=True)
```

```python
C5_Emean=C5.mean(dim='number',keepdims=True)
```
