```python
%%time
for mm in range(1,2):#13): # month loop
    print(f"Month:{mm} {dt.datetime.now()}",flush=True)
    fnamelast=fnameCanESMAnom(mdirC5,climyrs[0],climyrs[-1],climyrs[-1],mm) # file name associated with end of climatology, may not be true last but close
    if not os.path.exists(fnamelast): # skip if file already exists
        flist=[fnameCanESMdaily(mdirC5,yyyy,mm,1,0) for yyyy in range(1993,2025) if yyyy<2024 or mm<=6] # stop at Jul 2024
        # chunks={'reftime':-1,'leadtime':5,'r':-1,'lat':-1,'lon':-1}
        fnameclim=fnameCanESMClim(mdirC5,climyrs[0],climyrs[-1],mm)
        with xr.open_mfdataset(flist,parallel=True,combine='nested',concat_dim='reftime',
                               chunks={'reftime':-1,'leadtime':-1,'r':-1,'lat':-1,'lon':-1}) as ff:
            #display(ff)
            print('1')
            if os.path.exists(fnameclim):
                fclim=xr.open_dataset(fnameclim)
                EClim=fclim['tso']
            else:
                EClim=ff.tso.sel(reftime=slice(np.datetime64(f'{climyrs[0]:04}-01-01'),np.datetime64(f'{climyrs[-1]:04}-12-31'))).mean(dim='r').mean(dim='reftime').persist()
                EClim.to_netcdf(fnameclim,mode='w')
            print('A')
            Anom0=ff.tso-EClim
            for iy in range(1993,2025):
                if iy<2024 or mm<=6: # stop at Jul 2024
                    fname=fnameCanESMAnom(mdirC5,climyrs[0],climyrs[-1],iy,mm)
                    print('B')
                    Anom0=ff.tso.sel(reftime=np.datetime64(f'{iy:04}-{mm:02}-01'))-EClim
                    Anom0.rename('sst_an').to_netcdf(fname,mode='w')
            del EClim; del Anom0;
            try:
                fclim.close()
            except:
                pass
```
